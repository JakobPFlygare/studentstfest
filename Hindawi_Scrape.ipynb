```python
# -*- coding: utf-8 -*-
"""
Created on Fri Apr  5 16:36:40 2019
@author: jflygare
"""

from bs4 import BeautifulSoup
import requests 
import csv
import pandas as pd
import re
```

```python
def cleanText (strDirty):
    clean = ''.join([i for i in strDirty if not i.isdigit()])
    clean = re.sub("[!@#$+%*:()'-?°‘’]", '', clean)
    clean = re.sub(' +',' ',clean)
    return clean
```

```python
master_source = requests.get('https://www.hindawi.com/search/title/global+warming/or/abstract/global+warming/').content
url = 'https://www.hindawi.com/'
```

```python
master_soup = BeautifulSoup(master_source,'lxml')
    
#print(master_soup.prettify())

master_article = master_soup.find('div', id="SearchResultPnl")
#print(master_article.prettify())
```

```python
# Setting up csv files
csv_file = open('Hindawi_scrape.csv','w',encoding="utf-8")
csv_writer = csv.writer(csv_file)   
csv_writer.writerow(['title','abstract'])

for page in range(2,7):
    master_source = requests.get('https://www.hindawi.com/search/title/global+warming/or/abstract/global+warming/'+str(page)+'/').content
    master_soup = BeautifulSoup(master_source,'lxml')
    master_article = master_soup.find('div', id="SearchResultPnl")


    #Find title and abstract for all articles
    for article in master_article.find_all('li'):
        link = article.a['href']
        source = requests.get(url+link).content
        soup = BeautifulSoup(source,'lxml')

        article = soup.find('div',class_ = "rprt_all")

        title = soup.find('h2').text
        cleanTitle = cleanText(title)
        abstract = soup.find_all('div', class_='xml-content')[1].p.text
        cleanAbstract = cleanText(abstract)
        csv_writer.writerow([cleanTitle,cleanAbstract])


csv_file.close()
   
df = pd.read_csv('Hindawi_scrape.csv',sep=",")
#Looping into all pages
```
